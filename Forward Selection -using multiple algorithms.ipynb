```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
dataset1=pd.read_csv("prep.csv",index_col=None)
```

```python
df2=dataset1

df2 = pd.get_dummies(df2, drop_first=True)
```

```python
df2
```

```python
indep_X=df2.drop('classification_yes', 1)#drop output column separately using drop_function
dep_Y=df2['classification_yes']
```

```python
indep_X
```

```python
dep_Y
```

```python
algorithms=[LinearRegression,Ridge,Lasso,RandomForestRegressor]
```

```python
### Forward Selection:

def forward_selection(indep_X,dep_Y,algorithms):
    
    results=pd.DataFrame(columns=['Algorithm','Selected_Features','MSE'])
    
    for algorithm in algorithms:
        selected_features = []
        algorithm_name=algorithm.__class__.__name__
        
        # Loop through features and add the one that improves the model the most
        while len(selected_features)<len(indep_X.columns):
            best_feature = None
            best_mse = float('inf') 
            # Initialize with a large value

            for feature in indep_X.columns:
                if feature not in selected_features:
                    current_features = selected_features + [feature]
                    X_train, X_test, y_train, y_test = train_test_split(indep_X, dep_Y, test_size = 0.25, random_state = 0)

                    X_train_current = X_train[current_features]
                    X_test_current = X_test[current_features]


                    model= algorithm()
                    model1= model.fit(X_train_current, y_train)
                    predictions = model1.predict(X_test_current)
                    mse = mean_squared_error(y_test, predictions)

                    if mse < best_mse:
                        best_mse = mse
                        best_feature = feature

            # Add the best feature to the selected features
            selected_features.append(best_feature)
            #print("Selected features:",selected_features)   

            results=results.append({'Algorithm':algorithm_name,'Selected_Features':selected_features,'MSE':best_mse},ignore_index=True)
    return results
```

```python
results=forward_selection(indep_X,dep_Y,algorithms)
```
