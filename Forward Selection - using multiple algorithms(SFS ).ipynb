```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from mlxtend.feature_selection import SequentialFeatureSelector
from sklearn.metrics import mean_squared_error
```

```python
dataset1=pd.read_csv("House_Prediction.csv",index_col=None)
```

```python
df2=dataset1

df2 = pd.get_dummies(df2, drop_first=True)
```

```python
indep_X=df2.drop('price', axis=1)#drop output column separately using drop_function
dep_Y=df2['price']
```

```python
indep_X
```

```python
dep_Y
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(indep_X, dep_Y, test_size=0.2, random_state=42)
```

```python
# Initialize regression models
regressors = {
    'Decision Tree': DecisionTreeRegressor(random_state=42),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
    'Linear Regression': LinearRegression(),
    'SVM Linear': SVR(kernel='linear'),
    'SVM Non-linear': SVR(kernel='rbf')
}
```

```python
# Perform feature selection and regression for each algorithm
for algorithm, regressor in regressors.items():
    # Initialize SequentialFeatureSelector with forward selection
    sfs = SequentialFeatureSelector(regressor, k_features='best', forward=True, scoring='neg_mean_squared_error', cv=5)

    # Fit the SequentialFeatureSelector to the training data
    forward_feature_selection=sfs.fit(X_train, y_train)
```

```python
# selected indexes after the sfs.fit executed ,use this code to see the indexes
    selected_features_idx = list(forward_feature_selection.k_feature_idx_) 
```

```python
# Get the selected features(after the model fit using this below code to see the selected features)
    selected_features = list(forward_feature_selection.k_feature_names_) 
```

```python
# Get the the final mse score
selected_features_score = list(forward_feature_selection.k_score_)
```

```python
# in this dataframe to call dictionary_method(pd.datframe.from_dict after we pass the below agrument/paramenter in the dic function()then we cal transpose-T )
results=pd.DataFrame.from_dict(forward_feature_selection.get_metric_dict()).T
```
